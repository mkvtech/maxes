```python
%autoreload
from maxes.analyze_sequence import draw_nx_graph
import copy

def draw_nx_graph_aux(graph: nx.DiGraph, fig=None, ax=None, size_x=16, size_y=16, pos=None, shorten_node_names=True):
    if ax is None or fig is None:
        fig, ax = plt.subplots()

    fig.set_size_inches(size_x, size_y)

    if shorten_node_names:
        # graph = copy.deepcopy(shorten_node_names)
        mapping = {node: i for i, node in enumerate(graph.nodes)}
        graph = nx.relabel_nodes(graph, mapping)

    if pos is None:
        pos = nx.kamada_kawai_layout(graph)

    draw_nx_graph(graph, pos=pos, ax=ax)

draw_nx_graph_aux(generator.sequence_graph_, shorten_node_names=True)
```
