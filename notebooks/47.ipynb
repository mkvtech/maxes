```python
# Combine and save to file

# All in one
meta = {
    "Activities": "../output/experiment-46-4_activities_edited_hh104_labour.csv",
    "BPI Challenge 2020": "../output/experiment-46-4_bpic2020__request_for_payment.csv",
    "Alarm System": "../output/experiment-46-4_software_data_analytics__alarm_system.csv",
    "Photo Copier": "../output/experiment-46-4_photo_copier.csv",
    "CCC19": "../output/experiment-46-3_ccc19.csv",
    "Simple": "../output/experiment-46-4_simple.csv"
}

results = []
for i, (title, file_path) in enumerate(meta.items()):
    df = pd.read_csv(file_path)

    for name, group in df.drop(columns=["file_key"]).groupby("traces_count"):
        results.append({
            "title": title,
            "file_path": file_path,
            "traces_count": group["traces_count"].mean(),
            "percentage_of_fitting_traces": group['percentage_of_fitting_traces'].mean() / 100,
            "average_trace_fitness": group['average_trace_fitness'].mean(),
            "generalization": group['generalization'].mean(),
            "precision": group['precision'].mean(),
        })

df = pd.DataFrame(results)
df.to_csv("../output/experiments-1-traces-counts.csv")
```
